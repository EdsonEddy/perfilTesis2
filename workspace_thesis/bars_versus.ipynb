```python
import numpy as np
import matplotlib.pyplot as plt
 
serie_1 = [100,100,100,76.92,60.46,52.5,78.57,56.00]
serie_2 = [100,93.75,100,92.31,73.24,72.73,88.89,71.11]
serie_3 = [0,0,0,75,54.55,60,66.67,30]
 
numero_de_grupos = len(serie_1)
indice_barras = np.arange(numero_de_grupos)
ancho_barras =0.25
 
plt.bar(indice_barras, serie_1, width=ancho_barras, label='SCED', color='black')
plt.bar(indice_barras + ancho_barras, serie_2, width=ancho_barras, label='RKR-GST', color="gray")
plt.bar(indice_barras + ancho_barras + ancho_barras, serie_3, width=ancho_barras, label='WIN-FIN', color="silver")
plt.legend(loc='best', prop={'size': 8})
## Se colocan los indicadores en el eje x
plt.xticks(indice_barras + ancho_barras, [i+1 for i in range(len(serie_1))])
 
plt.ylabel('Porcentaje de similitud')
plt.xlabel('Conjunto')
#plt.title('Precision de los algoritmos')

target_path = "C:\\Users\\ACCESO\\Desktop\\experiments_file\\images\\results_PAIR.png"
plt.savefig(target_path, dpi=300)
plt.show()
```
